```python
x = np.arange(10)
y = [1,4,5,8,9,5,3,-2,-1,0]
plt.xlim(-2,10)
plt.ylim(-5,12)
plt.xticks([0,3,6])
plt.yticks([1,5,9])
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
x=[1,2,3,4]
y=[2,4,6,8]
z=[3,6,9,12]
plt.plot(x, label='x')
plt.plot(y, label='y')
plt.plot(z ,label='z')
plt.legend()
plt.show()
```

```python
data=pd.read_csv("Traffic_Accident_2017.csv", encoding="euc-kr")
```

```python
data.head()
```

```python
temp=data['요일'].value_counts()
```

```python
temp
```

```python
y=temp[["월","화","수","목","금","토","일"]]
```

```python
y
```

```python
x=range(7)
plt.bar(x,y)
plt.xticks(x,["월","화","수","목","금","토","일"])
plt.show()
```

```python
x=range(7)
plt.bar(x,y)
plt.xticks(x,["월","화","수","목","금","토","일"])
plt.ylim(500,620)
plt.xlabel("요일")
plt.ylabel("사고건수")
plt.title("요일별 사망교통사고 건수")
plt.show()
```
