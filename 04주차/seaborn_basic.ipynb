```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sb
```

```python
x=np.linspace(0,14,100)
y1=np.sin(x)
y2=2*np.sin(x+0.5)
y3=3*np.sin(x+1.0)
y4=4*np.sin(x+1.5)

plt.figure(figsize=(10,6))
plt.plot(x,y1,x,y2,x,y3,x,y4)
```

```python
sb.set_style('whitegrid')

plt.figure(figsize=(10,6))
plt.plot(x,y1,x,y2,x,y3,x,y4)
plt.show()
```

```python
#tips : 요일별 점심 저녁 흡연 여부와 식사 금액과 팁을 정리한 데이터 
tips=sb.load_dataset('tips')
tips.head()
```

```python
plt.figure(figsize=(8,6))
sb.boxplot(x='day',y='total_bill', data=tips)
plt.show()
```

```python
#hue 인수에 카테고리 변수 이름을 지정하여 카테고리 값에 따라 색상을 다르게 할 수 있다.
plt.figure(figsize=(8,6))
sb.boxplot(x='day',y='total_bill', hue='smoker',data=tips , palette='Set3')
plt.show()
```

```python
sb.set_style('darkgrid')
sb.lmplot(x='total_bill', y='tip',data=tips,size=7)
plt.show()
```

```python
sb.lmplot(x='total_bill', y='tip',hue='smoker',data=tips,palette='Set1',size=7)
plt.show()
```

```python
flights=sb.load_dataset('flights')
flights.head(5)
```

```python
#연도별 월별 항공기 승객수 기록한 데이터 
flights=flights.pivot("month","year","passengers")
flights.head(5)
```

```python
plt.figure(figsize=(10,8))
sb.heatmap(flights, annot=True, fmt='d') 
#annot : 선택 사항True이면 각 셀에 데이터 값을 씁니다 fmt :  문자열 서식 코드 .2f 실수형 
plt.show()
```
